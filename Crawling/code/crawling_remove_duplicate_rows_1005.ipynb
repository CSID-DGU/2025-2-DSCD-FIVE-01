```python
import pandas as pd

file_path = '/Users/zoohunn/Desktop/4-2/데사캡디/크롤링/unipass_crawling_0929.csv'

data = pd.read_csv(file_path, encoding='utf-8-sig', engine='python')
pd.set_option('display.max_colwidth', None)

print(data.shape)
data.head()
```

```python
# 중복 판단 기준 열 지정
cols = ['HSCode', '이름', '설명', '사유']

# 중복 개수 확인
dup_count = data.duplicated(subset=cols).sum()
print(f"중복된 행 수: {dup_count}")
```

```python
data[data.duplicated(subset=['HSCode', '이름'], keep=False)] 
```

```python
# 중복 제거
cleaned_data = data.drop_duplicates(subset=cols, keep='first').reset_index(drop=True)
print(f"제거 후 행 개수: {len(cleaned_data)} / 원본 행 개수: {len(data)}")                      # 약 30000개 확보
cleaned_data.head()                             
```

```python
print(cleaned_data.shape)
dup_count1 = cleaned_data.duplicated(subset=cols).sum()
print(f"중복된 행 수: {dup_count1}")
```

```python
# 필요 없는 열(페이지, 행 등) 제거
drop_cols = ['페이지', '행']
cleaned_data = cleaned_data.drop(columns=[c for c in drop_cols if c in cleaned_data.columns])

# 날짜형 변환 + 최근 날짜순 정렬
cleaned_data['시행일자'] = pd.to_datetime(cleaned_data['시행일자'], errors='coerce')
cleaned_data = cleaned_data.sort_values(by='시행일자', ascending=False, kind='stable')

# 새 번호 열 추가 (1부터 시작)
cleaned_data.insert(0, '번호', range(1, len(cleaned_data) + 1))
```

```python
# 저장
output_path = '/Users/zoohunn/Desktop/4-2/데사캡디/크롤링/unipass_crawling_1005_dedup.csv'
cleaned_data.to_csv(output_path, index=False, encoding='utf-8-sig')
```
